````python
import matplotlib.pyplot as plt
def classify_voltage_logic(volt):
    """Categorizes a single reading."""
    if 110 <= volt <= 130:
        return "Target"
    elif 0 <= volt < 110:
        return "Low"
    elif volt > 130:
        return "High"
    else:
        return "Anomaly/Error" # Handles negative values and other unclassified values

def category_summary(results):
    """Summarizes the counts of each category from a list of classification results."""
    summary = {'Low': 0, 'High': 0, 'Anomaly/Error': 0, 'Target': 0}
    for category in results:
        summary[category] += 1
    return summary

def generate_voltage_report(readings):
    # 1. Iniciar os contadores
    count_low = 0
    count_target = 0
    count_high = 0
    count_anomalies = 0

    for volt in readings:
        # Using the classify_voltage_logic to determine the category
        category = classify_voltage_logic(volt)
        if category == "Target":
            count_target += 1
        elif category == "Low":
            count_low += 1
        elif category == "High":
            count_high += 1
        elif category == "Anomaly/Error":
            count_anomalies += 1

    # 2. Imprimir o resumo
    print("===== RELATÓRIO FINAL =====")
    print(f"Target: {count_target}")
    print(f"Low: {count_low}")
    print(f"High: {count_high}")
    print(f"Anomalies/Errors: {count_anomalies}")
    print("===========================")
    return {'Low': count_low, 'High': count_high, 'Anomaly/Error': count_anomalies, 'Target': count_target}

# calling function generate Report
voltage_readings = [100, 115, 120, 130, 157, -32, 0, 82, 24, 500, 127, 129.1, 143.2]
report_summary = generate_voltage_report(voltage_readings)
# Create a new list using your logic
results = []
for v in voltage_readings:
    category = classify_voltage_logic(v)
    results.append(category)

print(results)
summary_counts = category_summary(results)
print(summary_counts)

# 1. Prepare the data from your summary_counts dictionary
categories = list(summary_counts.keys())
values = list(summary_counts.values())

# 2. Create the plot
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=['blue', 'red', 'gray', 'green'])

# 3. Add professional labels
plt.title('Voltage Reading Distribution Report',fontsize=14)
plt.xlabel('Category', fontsize=14 )
plt.ylabel('Number of Readings', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 4. Show the chart
plt.show()
````

````python
import pandas as pd

def deviation(volt):
  # The function should accept a single voltage value and return its deviation
  return volt - 120

# 1. Create a dictionary with your raw data
data = {
    'sensor_id': ['SN-001', 'SN-002', 'SN-003', 'SN-004', 'SN-005', 'SN-006', 'SN-007', 'SN-008', 'SN-009', 'SN-010', 'SN-011', 'SN-012', 'SN-013'],
    'voltage_reading': [100, 115, 120, 130, 157, -32, 0, 82, 24, 500, 127, 129.1, 143.2]
}

# 2. Convert to a DataFrame (Table)
df = pd.DataFrame(data)

# 3. Apply your logic to create a NEW column
def classify_voltage_logic(volt):
    """Categorizes a single reading."""
    if 110 <= volt <= 130:
        return "Target"
    elif 0 <= volt < 110:
        return "Low"
    elif volt > 130:
        return "High"
    else:
        return "Anomaly/Error" # Handles negative values and other unclassified values

# We use .apply() to run your function on every row of the 'voltage_reading' column
df['status'] = df['voltage_reading'].apply(classify_voltage_logic)
df['deviation'] = df['voltage_reading'].apply(deviation)
# 4. View the results
print("First 5 rows of your new Dataset:")
print(df.head())

# 5. Get the summary instantly with value_counts()
print("\nQuick Summary:")
print(df['status'].value_counts())

# Filter the table to see ONLY the problems
high_voltage_sensors = df[df['status'] == 'High']

print("--- Maintenance Required for these Sensors ---")
print(high_voltage_sensors)

# Calculate the average of ONLY the target readings
avg_target = df[df['status'] == 'Target']['voltage_reading'].mean()
print(f"\nAverage Target Voltage: {avg_target:.2f}V")
````

````python
import numpy as np
import pandas as pd
# 1. We create a "Dirty" version of your sensor data
dirty_data = {
    'sensor_id': ['SN-01', 'SN-02', 'SN-03', 'SN-04', 'SN-05'],
    'voltage': [120, np.nan, 130, 'error', 115] # Notice the NaN and the string 'error'
}
df_dirty = pd.DataFrame(dirty_data)

print("--- The Dirty Data ---",'\n',df_dirty)


# 2. THE MISSION: Clean it!
# Convert 'voltage' to numeric, turning 'error' into NaN automatically
df_dirty['voltage'] = pd.to_numeric(df_dirty['voltage'], errors='coerce')

# Drop rows that have NaN (Missing values)
df_clean = df_dirty.dropna(subset=['voltage'])

print("\n--- The Cleaned Data ---",'\n',df_clean)
````

````python
import numpy as np
import pandas as pd

# 1. Create a "Dirty" dataset
raw_data = {
    'sensor_id': ['SN-01', 'SN-02', 'SN-03', 'SN-04'],
    'reading': [120, 'ERR', 115, np.nan] # Notice the string 'ERR' and the missing value np.nan
}
df_dirty = pd.DataFrame(raw_data)

# 2. Convert everything to numbers (This turns 'ERR' into NaN)
df_dirty['reading'] = pd.to_numeric(df_dirty['reading'], errors='coerce')

# 3. Remove the rows that have missing values (NaN)
df_clean = df_dirty.dropna(subset=['reading'])

print("Cleaned Data:",'\n',df_clean)
````
